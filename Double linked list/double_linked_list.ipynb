```python
#Node ou nó
class Node:
  def __init__(self, label):
    self.prev = None
    self.label = label
    self.next = None
    self.body_removed = False

  def getPrev(self):
    return self.prev

  def setPrev(self, prev):
    self.prev = prev

  def getLabel(self):
    return self.label

  def setLabel(self, label):
    self.label = label

  def getNext(self):
    return self.next

  def setNext(self, next):
    self.next = next

class DoubleLinkedList:

  def __init__(self, size):
    self.first = None
    self.len_list = 0
    self.size = size

# Para o "ADD"
  def push(self, label): #inserir ordenado    
    node = Node(label) #cria o novo nó

    # Lista vazia, simplesmente faz a LinkedList apontar para o novo node
    if self.empty():
      self.first = node

    # Lista não vazia
    else:
      if label < self.first.getLabel():
        #inserção no começo
        node.setNext(self.first) # node apontar  para  o começo da lista
        self.first.setPrev(node) # Mudança devido a LISTA DUPLAMENTE ENCADEADA
        self.first = node # LinkedList apontar para o novo node 

      else:
        # Inserir no meio ou no final
        aux1 = self.first
        aux2 = self.first.getNext()
        flag_insert = False
        while aux2 != None:
          if label < aux2.getLabel():
            # Inserir no meio
            node.setPrev(aux1) # Mudança devido a LISTA DUPLAMENTE ENCADEADA
            node.setNext(aux1.getNext()) #node.setNext(aux2) 
            aux1.setNext(node)
            aux2.setPrev(node) # Mudança devido a LISTA DUPLAMENTE ENCADEADA
            flag_insert = True
            break
          aux1 = aux1.getNext()
          aux2 = aux2.getNext()
        #inserir no final
        if flag_insert == False:
          aux1.setNext(node) 
          node.setPrev(aux1) # Mudança devido a LISTA DUPLAMENTE ENCADEADA      
    self.len_list += 1
       
  def pop(self, label):
    # Não remover de uma lista vazia  
    if not self.empty():
      flag_remove = False

      # Remover de uma lista com apenas 1 elemento
      if self.first.getNext() == None and self.first.getLabel() == label:
        self.first = None
        flag_remove = True

      # Remover o primeiro node da lista
      elif self.first.getLabel() == label:
        self.first = self.first.getNext()
        self.first.setPrev(None) # Mudança devido a LISTA DUPLAMENTE ENCADEADA
        flag_remove = True

      # Remover do meio ou final da lista
      else:
        aux1 = self.first
        aux2 = self.first.getNext()
        while aux2 != None:
          if aux2.getLabel() == label:
            aux1.setNext(aux2.getNext())
            if aux2.getNext() != None:
              aux2 = aux2.getNext() # Mudança devido a LISTA DUPLAMENTE ENCADEADA
              aux2.setPrev(aux1) # ou em 1 linha aux2.getNext().setPrev(aux1) # Mudança devido a LISTA DUPLAMENTE ENCADEADA
            flag_remove = True
            break
          aux1 = aux1.getNext()
          aux2 = aux2.getNext()

      if flag_remove == True:
        self.len_list -= 1
      else:
        print("Label não encontrada")

# Para o "REM"
  def remove_body(self, label):
    flag = False # Flag vai me retornar True = Se for para eu printar o valor da outra var
                 # Ou vai me retornar False = para eu retornar um ?.?

    if not self.empty():
      # Remover de uma lista com apenas 1 elemento
      if self.first.getNext() == None or self.first.getLabel() == label:
        self.first.body_removed = True
        return True, '0'        

      # Remover do meio ou final da lista
      else:
        aux1 = self.first # Node
        cont = 0
        while aux1 != None:
          if aux1.getLabel() == label:
            aux1.body_removed = True
            flag = True
            return flag, str(cont)
          aux1 = aux1.getNext()
          cont = cont + 1

        if flag == False:
          return flag, str(cont)
    
    else:
      print("Lista vazia - Função remove_body")

# Para o "QRY"
  def locate_num(self, label):
    flag = False # Flag vai me retornar True = Se for para eu printar o valor da outra var
                 # Ou vai me retornar False = para eu retornar um ?.?
                 
    if not self.empty():
      # Remover de uma lista com apenas 1 elemento
      # MUDEEEEEEEEEEEEIIIIIIIIII AQUUUUUUUUUUUUI, ANTES ERA OR
      if self.first.getNext() == None or self.first.getLabel() == label:
        flag = True
        return flag, '0'

      # Remover do meio ou final da lista
      else:
        aux1 = self.first # Node
        cont = 0
        while aux1 != None:
          if aux1.getLabel() == label:
            flag = True
            return flag, str(cont)
          aux1 = aux1.getNext()
          cont = cont + 1

        # TODO: Preciso achar uma forma de printar isso, se eu não achar nada vou precisar printar isso, mas o código coloca um print(f"{andar}." + locate_num(label)")
        if flag == False:
          return flag, str(cont)

    else:
        print("Lista vazia da função 'locate_num(self, label)'")

  def redistribuir(self, novo_predio):
    self.novo_predio = novo_predio # Novo prédio, com mais andares, ta vazio.
    
    

    if not self.empty():
      aux = self.first

      while aux.getNext() != None:
        print(aux.getLabel(), end=' -> ')
        aux = aux.getNext()
      print(aux.getLabel(), end=' ')
      print('')

      print('\n')

  def empty (self):
    if self.len_list == 0:
    # if self.first == None
      return True
    return False

  def length(self):
    return self.len_list

  def show(self):
    if not self.empty():
      aux = self.first

      print("Ordem crescente: ")
      while aux.getNext() != None:
        print(aux.getLabel(), end=' -> ')
        aux = aux.getNext()
      print(aux.getLabel(), end=' ')
      print('')

      print('\n')

      # Percorrendo na ordem inversa
      # print("Ordem decrescente: ")
      # while aux.getPrev() != None:
      #   print(aux.getLabel(), end=' -> ')
      #   aux = aux.getPrev()      
      # print(aux.getLabel(), end=' ')  
      # print('')

  def clear(self):
        self.first = None
        self.len_list = 0
```

##### A estrutura de jazigos

Em cada andar teremos uma distribuição dessa.

Então por exemplo: 

input[0] = M L

M -> Número inicial de andares, será o meu andares.len() = M </br>
L -> Número de jazigos por andar do cemitério

Logo depois virá uma sequência de instruções "ADD", "REM" e "QRY" seguido por "K" (o número do seguro social).</br>

* Para determinar o andar que acessaremos para alocar o corpo faremos um M % K == M mod K -> Isso resultará no número</br>
do andar que deveremos acessar para alocar o corpo no jazido daquele andar.</br>

* Cada andar terá a estrutura de double linked list, com um lenght máximo de L.</br>

* Se não ouver mais espaços disponíveis no andar que aquele corpo deveria entrar faremos o insert do corpo no próximo andar,</br>
obedencendo sempre a mesma regra, em ordem crescente, se não tiver mais espaço avançar para o próximo andar.</br>

* SE NÃO HOUVER MAIS JAZIDOS DISPONÍVEIS NAQUELE CEMITÉRIO, A LEI LOCAL MANDA QUE UM NOVO PRÉDIO DE 2M + 1 ANDARES,</br>
COM OS MESMOS L TÚMULOS POR ANDAR, SEJA IMEDIATAMENTE CONSTRUIDO NO LOCAL. OS TUMULOS VAZIOS SÃO IGNORADOS, OU SEJA,</br>
